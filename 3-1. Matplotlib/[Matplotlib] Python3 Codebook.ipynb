## 🔎 Basic Methods

<h2>🎈 모든 그래프는 한 figure 안에 여러 개 추가가능하다!!</h2>

 - **plt .show ( )**: 그래프 보기
 - **plt .subplot ( a , b , c )**: a*b matrix로 figure을 나누고 그 중에 c에 plot을 그리겠다는 의미!
 - **plt .subplots ( rows , cols , sharex=False, sharey=False, squeeze=True, subplot_kw=None, gridspec_kw=None, fig_kw)**: 사용하는 parameter는 plt.subplot()과 같고, return값이 'fig'와 'ax'로 두 개의 값을 가진다.
   - sharex: x축을 공유
   - sharey: y축을 공유
   - fig: Figure 
   - ax: Object or Array of Axes objects
 - **plt .add_subplot ( a , b , c )**: a*b matrix로 figure을 나눈 곳에서 c에 plot 추가
   - **.set_xlim ( min , max )**: min~max의 xlim을 설정해줄 수 있음!!
 - **plt .clf( )**: 그림 clear하기
 - **plt .xscale ( ' log ' )** : log scale 씌우기
 - **plt .savefig ( " .png " )**: 그림 저장하기 

<h2>🎈 plt.plot ( ) </h2>

- 가장 기본적인 형태도 ( )안에 리스트, array 형식의 자료형을 넣을 수 있음
- **plt .plot (x)**: x로만 그래프 그리기
- **plt .plot ( x , y , " line type과 color " )**: x, y 데이터로 그래프 customizing하기
- **plt .plot ( x, y , color = " ", linestyle = " ", label = " 레이블 이름 " )**: 그리려는 두 변수의 길이가 같아야 함
- **plt .plot ( x, y, "line type과 color", linewidth = 두께 , markersize = 크기, label = " 레이블 이름 " )**

<h2>🎈 plt.figure ( ) </h2>

- 그림 그릴 판을 설정해주는 것
- **plt .figure ( figsize = ( a , b ), facecolor = " 배경색 " )**: 가로 a, 세로 b과 배경색을 설정가능
  - **.subplot2grid( ( shape , shape ) , ( loc , loc ) , rowspan = , colspan = )**: 크기를 조절 가능!


<h2>🎈 plt.label, plt.title, plt.legend</h2>

- 레이블, 제목, 레전드 만들기!
- **plt .xlabel ( ' 레이블 이름 ' , fontsize = , labelpad = , rotation = 회전정도 )** : x label 달기
- **plt .ylabel ( ' 레이블 이름 ' , fontsize = , labelpad = , rotation = 회전정도 )** : y label 달기
- **plt .title ( ' 타이틀 이름 ', fontsize = , pad = , loc = 위치 )** : title 달기
- **plt .legend ( )**: label을 보여줌. loc을 통해서 legend의 위치도 조정 가능

  **plt .legend ( loc = , facecolor = " " , edgecolor = " ", framealpha = )**
  - 0 best
  - 1 upper right
  - 2 upper left
  - 3 lower left
  - 4 lower right
  - 5 right
  - 6 center left
  - 7 cener right
  - 8 lower center
  - 9 upper center
  - 10 center

<h2>🎈 plt.ticks </h2>

- x, y축의 범위가 아닌 눈금을 설정해주는 것임!!
- **plt .xticks ( [ Tick location ] , [ Tick Label ] , color = " " )**: color말고 다른 parameters도 있음
- **plt .yticks ( [ Tick location ] , [ Tick Label ] , color = " " )**: color말고 다른 parameters도 있음
> 밑에 선배가 만들어놓은 함수 있음~ 참고~


<h2>🎈 plt.gca() </h2>

- 해당 fig의 axes instance를 return
- **plt .gca ( ) .set_facecolor( " ")**: 배경색 설정해주기
- **plt .gca ( ) .spines [ ' ' ] .set_bounds ( 시작 , 끝  )**
- **plt .gca ( ) .spines [ ' ' ] .set_color (' ')**
- **plt .gca ( ) .spines [ ' ' ] .set_linewidth ( )**
- **plt .gca ( ) .spines [ ' ' ] .set_position ( )**


<h2>🎈 plt.grid, plt.lim </h2>

- grid와 x, y축의 범위를 설정해주기
- **plt .grid (True, linewidth = , linestyle = " ", color = " ")**: grid 설정해주기
- **plt .xlim ( 최소, 최대 )**: x축의 범위를 설정해주기
- **plt .ylim ( 최소, 최대 )**: y축의 범위를 설정해주기


<h2>🎈 plt.text, plt.annotate </h2>

- 글자 넣기
- **plt .text ( x = 위치 , y = 위치 , s = ' 내용 쓰셔유 ' , color = ' 색 ', weight = ' 설정 ' , rotation = 회전, backgroundcolor = ' ', alpha = 투명도, size = )**: 이렇게 글씨를 넣을 수 있음
- **plt .annotate ( 넣고 싶은 text, xy = ( , ) , xycoords = ' ' , xytext = ( , ), textcoords = ' ' , fontsize = , arrowprops = dict ( arrowstyle = " " , connectionstyle = " ")**: 이렇게 하면 특정값에 text를 넣을 수 있음
  - xytext: 텍스트의 위치
  - xy: 화살표의 목적지
  - arrowprops: 화살표 설정
  - xycoords: data가 default
    - figure points, figure pixels, figure fraction, axes points, axes pixels, axes fraction, polar가 올 수 있음
  - textcoords: offset points, offset pixels 2개 조건이 있다. 쓰면 pixel 밖으로 arrow가 넘어가지 않는다..

````python
# .fill_between()
def graph_data():
    
    fig=plt.figure()
    ax1=plt.subplot2grid((1,1),(0,0))
   
    plt.ylabel('Price')#이름 붙이기
    plt.xlabel('Year')
    
    print('Currently viewing: replacement data')
    url = 'https://pythonprogramming.net/yahoo_finance_replacement'
    source_code = urllib.request.urlopen(url).read().decode()
    stock_data = []
    split_source = source_code.split('\n')
    for each_line in split_source:
        split_line = each_line.split(',')
        #if len(split_line) == 7:
        if 'Date' not in each_line:
            stock_data.append(each_line) 
                           
    date, openp, highp, lowp, closep, adjusted_close, volume = np.loadtxt(stock_data, delimiter = ',',
                                         unpack = True, converters={0: mdates.bytespdate2num('%Y-%m-%d')})                 
    print(date)
    plt.plot_date(date,closep,'-')
    
    ax1.fill_between(date,closep,200,where=(closep>=200),facecolor='g',alpha=0.5)#특정구간만 색칠하기
    ax1.fill_between(date,closep,100,where=(closep<=100),facecolor='r',alpha=0.5)#특정구간 색칠하기 2
    ax1.grid(True) #color='g',linestyle='-',linewidth='3')
    ax1.yaxis.label.set_color('m') #색변경
    ax1.xaxis.label.set_color('c')
    ax1.set_yticks([0,200,400,600,800])#범위변경
    
    ax1.spines['left'].set_color('c') #spine 색 변화
    ax1.spines['bottom'].set_color('c')
    
    ax1.spines['top'].set_visible(False)# 안 보이게 하기: 원래는 위에 라인이 있음
    ax1.spines['right'].set_visible(False)
    
    ax1.axhline(200,color='r')# 수평선 그리기
    ax1.axhline(300,color='g')
    
    ax1.spines['left'].set_linewidth(5)
    plt.subplots_adjust(left=.09,bottom=.16,right=.94,top=.95,wspace=.2,hspace=.2)
    plt.show()
 
#stock = input('Stock to plot: ')
graph_data()
````

````python
# tick 만드는 함수
## 선배가 만들어준 귀찮아서 만든 함수
## 라벨 넣으면, 라벨 수만큼 알아서 쪼개주고, 숫자 넣으면, 숫자대로 범위를 쪼갠다음 소숫점 아래 2번째 아래 까지 해당 값을 return
def t_generator(xy,x_lower,x_upper, labels):
    if type(labels) == int:
        label_length=labels-1
        tick_range=np.arange(x_lower,x_upper+0.00001,(x_upper-x_lower)/(label_length))
        labels=np.around(tick_range,2)
    else:
        label_length=(len(labels)-1)
        tick_range=np.arange(x_lower,x_upper+1,(x_upper-x_lower)/(label_length))
    if str(xy)=="x":
            return plt.xticks(tick_range,labels)
    if str(xy)=="y":
        return plt.yticks(tick_range,labels)
````

````python
# auto label 만드는 함수
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
````
